Hipotesis 4, standarScales y modelo RandomForest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

Usamos el standard scaler para estandarizar los valores restando la media y dividiendo por la desviación estándar.

```python
from sklearn.preprocessing import StandardScaler

# Seleccionar las columnas a escalar
columns_to_scale = ["sqft_living", "sqft_lot", "sqft_basement"]

# Inicializar el escalador
scaler = StandardScaler()

# Aplicar el escalado a las columnas seleccionadas
df2[columns_to_scale] = scaler.fit_transform(df2[columns_to_scale])
```

Usamos el modelo de RandomForest porque con la regresion lineal, no nos daba valores optimos, al usar este modelo podemos observas que el MSE ya es bastante optimo dando a entender que no solo hay que probar con un modelo, si no con el que mas se ajuste al ejercicio

En la grafica se observa que los valores estan muy cercanos se podria decir que el modelo podria predecir el valor de una vivienda

```python
X_hipot3 = df2.drop('price',axis=1) #  variables independientes
y_hipot3 = df2['price'] #  variable depnediente

# preparamos train data y test data
from sklearn.model_selection import train_test_split

X_hipot3_train, X_hipot3_test, y_hipot3_train, y_hipot3_test = train_test_split(X_hipot3, y_hipot3, test_size=0.30, random_state=42)

# Inicializar el modelo de Bosques Aleatorios
random_forest = RandomForestRegressor(n_estimators=100, random_state=42)

# Entrenar el modelo con las características y etiquetas de entrenamiento
random_forest.fit(X_hipot3_train, y_hipot3_train)

# Realizar predicciones en el conjunto de prueba
predictions = random_forest.predict(X_hipot3_test)

# Evaluar el rendimiento del modelo
mse = mean_squared_error(y_hipot3_test, predictions)
r2 = r2_score(y_hipot3_test, predictions)

print(f'Mean Squared Error: {mse}')

# Visualizar las predicciones vs. los valores reales
plt.scatter(y_hipot3_test, predictions)
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Predicciones vs. Valores Reales')
plt.show()
```
